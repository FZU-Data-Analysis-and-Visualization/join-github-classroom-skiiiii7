```python
import matplotlib.pyplot as plt
import numpy as np

# 生成一些数据
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 画图
plt.plot(x, y)
plt.title("测试绘图 - 正弦波")
plt.xlabel("X轴")
plt.ylabel("Y轴")

# 显示图像
plt.show()
```
