4. 新增一列 grade：score≥90 为优秀，否则为良好。

### 三、缺失值

1. 创建含缺失值的 DataFrame：

   data = {"x":[1, None, 3], "y":[None, 5, 6]}

   检测缺失值并统计每列缺失数量。

```python
import pandas as pd

data = {
    "x": [1, None, 3],
    "y": [None, 5, 6]
}

df = pd.DataFrame(data)

print("原始 DataFrame：")
print(df)

print("检测缺失值：")
print(df.isnull())

print("每列缺失值数量：")
print(df.isnull().sum())
```

2. 用 0 填充缺失值；删除含缺失值的行。

```python
import pandas as pd

data = {
    "x": [1, None, 3],
    "y": [None, 5, 6]
}

df = pd.DataFrame(data)

df_fill = df.fillna(0)

df_drop = df.dropna()

print("原始 DataFrame：")
print(df)

print("用 0 填充缺失值后：")
print(df_fill)

print("删除含缺失值的行后：")
print(df_drop)
```

### 四、分组聚合

1. 按 gender 分组，求 score 的均值和最大值。

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})

result = df.groupby("gender")["score"].agg(["mean", "max"])

print(result)
```

2. 按 gender 分组，统计每组人数。

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})

result = df.groupby("gender").size()

print(result)
```

### 五、排序与去重

1. 按 score 升序排序。

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})

result = df.sort_values(by="name")

print(result)
```

2. 给 DataFrame 添加一行重复数据（如 Tom,85, 男），去重。

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})

new_row = pd.DataFrame({
    "name": ["Tom"],
    "score": [85],
    "gender": ["男"]
})

df = pd.concat([df, new_row], ignore_index=True)

print("添加重复行后：")
print(df)

df = df.drop_duplicates()

print("去重后：")
print(df)
```

### 六、合并与拼接

1. 横向合并：

   df1 = pd.DataFrame ({"id":[1,2], "name":["X","Y"]})

   df2 = pd.DataFrame ({"id":[1,2], "age":[20,30]})

   按 id 合并。

```python
import pandas as pd

df1 = pd.DataFrame({"id": [1, 2], "name": ["X", "Y"]})
df2 = pd.DataFrame({"id": [1, 2], "age": [20, 30]})

result = pd.merge(df1, df2, on="id")

print(result)
```

2. 纵向拼接：

   df_a = pd.DataFrame ({"val":[1,2]})

   df_b = pd.DataFrame ({"val":[3,4]})

   拼接后重置索引。

```python
import pandas as pd

df_a = pd.DataFrame({"val": [1, 2]})
df_b = pd.DataFrame({"val": [3, 4]})

result = pd.concat([df_a, df_b], ignore_index=True)

print(result)
```

### 七、时间处理

1. 创建时间列：["2025-01-01","2025-01-02","2025-01-03"]，转为 datetime 格式。

```python
import pandas as pd

df = pd.DataFrame({
    "date": ["2025-01-01", "2025-01-02", "2025-01-03"]
})

df["date"] = pd.to_datetime(df["date"])

print(df)
print(df.dtypes)
```

2. 提取时间列的月份、星期。

```python
import pandas as pd

df = pd.DataFrame({
    "date": ["2025-01-01", "2025-01-02", "2025-01-03"]
})

df["date"] = pd.to_datetime(df["date"])

df["month"] = df["date"].dt.month
df["weekday"] = df["date"].dt.day_name()

print(df)
```

### 八、高级操作

1. 对 score 列做标准化（(值 - 均值)/ 标准差）。

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})

df["score_std"] = (df["score"] - df["score"].mean()) / df["score"].std()

print(df)
```

2. 用透视表：行是 gender，值是 score，聚合函数为中位数。

```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Tom", "Jerry", "Mike"],
    "score": [85, 92, 78],
    "gender": ["男", "女", "男"]
})

result = pd.pivot_table(
    df,
    index="gender",
    values="score",
    aggfunc="median"
)

print(result)
```
